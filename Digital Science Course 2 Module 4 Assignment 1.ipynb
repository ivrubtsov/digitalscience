### My code calculates the total amount of pull-ups if they are performed in a ladder-style. It means that I hit one rep then I have 30 seconds rest, then I do two reps, then three reps and so on. My code calculates the total amount of pull-ups for every try. It is limitid to 20 tries because I have never seen a person who can do more.

```python
total=0
for x in range(20):
    total+=x+1
    print("try "+str(x+1)+": "+str(total)+" total pull-ups")
```
